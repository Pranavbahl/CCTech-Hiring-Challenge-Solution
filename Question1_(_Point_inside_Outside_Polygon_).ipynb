```
# POINT INSIDE OR OUTSIDE POLYGON PROBLEM SOLUTION 

def rayTracing(x,y,shape):

    n = len(shape)
    k = False

    p1x,p1y = shape[0]
    for i in range(n+1):
        p2x,p2y = shape[i % n]
        if y > min(p1y,p2y):
            if y <= max(p1y,p2y):
                if x <= max(p1x,p2x):
                    if p1y != p2y:
                        Flag = (y-p1y)*(p2x-p1x)/(p2y-p1y)+p1x
                    if p1x == p2x or x <= Flag:
                        k = not k
        p1x,p1y = p2x,p2y

    return k


#EXAMPLE1
polygn = [[1, 0], [8, 3], [8, 8], [1, 5]]
points = [(3,5)]
kSharp = [rayTracing(point[0], point[1], polygn) for point in points]
print(kSharp)

#EXAMPLE2
polygn =[[-3,2], [-2,-0.8], [0,1.2], [2.2,0], [2,4.5]]
points = [(0,0)]
kSharp = [rayTracing(point[0], point[1], polygn) for point in points]
print(kSharp)
```
